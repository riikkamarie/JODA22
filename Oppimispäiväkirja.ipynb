### Data-analytiikka ja sen tasot 

Data-analytiikan osa-alueeseen olen ennestään perehtynyt aika paljonkin alan kirjallisuuden myötä, joten aihe luennolla oli todella mielekäs. Itselle analytiikan nelikentän "diagnosoiva analytiikka" oli kuitenkin vieraampi. Käsite on tullut vastaan, mutten ole itse käyttänyt sitä yhtenä analytiikan erillisenä osa-alueena. Alan kirjallisuudessa (Delen & Demirkan 2013; King 2016; Sharda et al. 2018, s. 49) on esitelty analytiikan jako vain kolmeen eri tasoon, joita ovat kuvaileva analytiikka, ennakoiva analytiikka ja ohjaava analytiikka, mutta nähtävästi yksiselitteistä jakotapaa ei ole luotu. En tiedä, onko automaattinen analytiikka terminä vielä kovinkaan vakiintunut, mutta kyseinen termi on tullut ainakin Davenportin & Harrisin artikkelissa (2017)vastaan analytiikan edistyneimpänä ja "korkeimpana" tasona. Fitz-enz ja Mattox, (2014) huomauttavat, että yrityksen saavuttaessa ohjaavan analytiikan tason, kasvaa analytiikan tuottama arvo eksponentiaalisesti kuvaavaan ja ennustavaan analytiikkaan verrattuna. Potentiaalisen kilpailuedun ja kyvykkäämmän analytiikan ratkaisumallien vaikutussuhdetta olen havainnollistanut alla olevassa kuvassa.

Ryömijät ja raapijat olivat ennalta täysin tuntematon aihealue. Tutustuin luennon myötä Scrapy-nimiseen työkaluun (raapija), joka on avoimen lähdekoodin ryömijä. Lisäksi tein pienen numerotesti-kokeilun havainnollistaaksen eri datatyyppejä (Lähde: luentomateriaalia mukaillen): 

```python
userInput=input('Anna numero välillä 0-5: ')
try:
    sumValue=1+int(userInput)
    print('Tulos:' - str(sumValue))
    
except ValueError:
    print('Et tainnut antaa numeroa?\n')
```

```python
userInput=input('Anna kokonaisluku välillä 0-5: ')
# userInput=int(userInput)
while userInput.isdigit()==False:
    userInput=input('Ei onnistunut. Anna kokonaisluku välillä 0-5: ')
try:
    sumValue=1+int(userInput)
    print('Lisäsin antamaasi lukuun yhden ja siitä tuli ' + str(sumValue)+".")
except TypeError:
    print('Et tainnut antaa numeroa?')
```

Yhteenvedossa käyn vielä läpi kuinka omasta mielestäni täytin oppimistavoitteet. Oppimista käyn läpi kokonaisuudessaa kurssin osalta sekä erikseen harjoitustyön näkökulmasta. Lopuksi haluan antaa vielä palautetta kurssista tämän oppimispäiväkirjan välityksellä. 

### Oma oppiminen suhteessa oppimistavoitteisiin

 #### Kokonaisuus 

Itseäni jäi hieman mietityttämään, pystyinkö tuomaan kaiken oppimani esille harjoitustyön ja oppimispäiväkirjan myötä. Aihepiiri oli itselleni aivan uusi ja tausta huomioiden sain lähteä aivan alkumereiltä kehittämään tietämystä datatieteen osalta. Käytin opintojakson osalta aikaa varmasti enemmän kuin muilla opintojaksoilla, koska halusin sisäistäväni asioita jatkoa ajatellen ja aihe kiinnostaa. Yleensä tavoittelen lähestulkoon tähtiä taivaalta, mutta nyt koitin pysyä perusasioissa, jotta sain luotua hyvän pohjan jatkoon ja tulevaan oppimiseen. Haasteita opintojakson ajankohdassa omalle kohdalleni toi ajallisesti haastava elämäntilanne, kun valmistuminen häämöttää ja diplomityön aikataulu on hurjan tiukka ja samalla opintovapaani työelämästäkin loppui ja palasin töihin kesken opintojakson. Koen kuitenkin, että omalle kohdalle asettamat tavoitteeni täyttyivät kurssin osalta ja mitä pidemmälle kurssi eteni, sitä vahvemmin aihe imaisi mukanaan, kun tuntui että alkoi oikeasti käsittämään mistä on kyse ja mitä hyötyä voin kurssin tiedoilla työlämässäkin saavuttaa. 

#### Harjoitustyö

Kokonaisuudessaan harjoitustyö oli todella aikaa vievä ja kärsivällisyyttä tarvittiin useaan otteeseen. Harjoitustyö oli kuitenkin myös todella opettavainen ja loppujen lopuksi voin sanoa sisäistäneeni paljon datatiedeprojektiin kuuluvia ominaispiirteitä ja tehtäviä. Tämän harjoitustyön osalta ei oikein voi antaa kattavia liiketoimintavinkkejä, mutta hyödyllisiä vinkkejä ja johtopäätöksiä työni avulla voi antaa. Harmillista oli myös huomata, kuinka moni muukin on tehnyt samankaltaista analysointia Airbnb-asuntoihin, mutta toisaalta tämä ei vähennä oman työn arvo tai oman oppimisen määrää. Valitsemallani datalla muuttujien korrelaatiot jäivät todella pieniksi, mikä vaikeutti myös mielenkiintoisten havaintojen tekemistä. Tämän harjoitustyön laajuuden puitteissa onnistuin kuitenkin mielestäni ylittämään omat oppimistavoitteeni ja koin harjoitustyön todella hyödylliseksi. Tästä on varmasti hyvä jatkaa ja kehittyä datatieteen parissa.

#### Palaute

Kaiken kaikkiaan pidin kurssista ja sen toteutustavasta. Harjoitustyö ja oppimispäiväkirja menetelminä olivat hyvä kombo, joiden myötä oppimiskokemus oli kattava niin teoreettisesti kuin käytännön tekemisen osaltakin. Luennot olivat mielenkiintoisia ja onneksi tallenteet olivat saatavilla, koska katsoin kaikki luennot tallenteiden kautta vaikka pari luentoa pääsinkin reaaliaikaisesti seuraamaan. Koodiklinikat olivat arvokkaita erityisesti harjoitustyön osalta ja pidin opintojakson asiantuntevasta mutta rennosta otteesta. Opintojakson toteuttamisen iso onnistuminen on tietysti aina se, että ainakin oma mielenkiintoni heräsi aihepiiriin, vaikka turhautumisen hetkiä ja kysymysmerkkejä itselläni matkan varrella hurja määrä olikin. Turhautumiset kuitenkin palkittiin ja kysymykset saivat vastauksensa! :)
